```python
import PIL.Image as pilimg
import numpy as np
import matplotlib.pyplot as plt
```

```python
im = pilimg.open( "./homework.JPG" )
```

```python
pix=np.array(im)
```

```python
pix
```

```python
pix.shape
```

```python
pix1=np.swapaxes(pix,0,1)
```

```python
plt.imshow(pix1)
```

```python
pix1=pix1[:193,:,:]
```

```python
plt.imshow(pix1)
```

```python
addline=np.ones(shape=(10,147,3),dtype=np.int8)
addline=addline*255
pix1=np.concatenate((pix1,addline),axis=0)
```

```python
plt.imshow(pix1)
```

```python
pix1.shape
```

```python
pix_r=pix1[:,:,0]
pix_g=pix1[:,:,1]
pix_b=pix1[:,:,2]
pix_g=(pix_r+pix_g+pix_b) / 3
```

```python
pix_g.shape
```

```python
pix_g1=pix_g[:,:,np.newaxis]
pix_g2=pix_g[:,:,np.newaxis]
pix_g3=pix_g[:,:,np.newaxis]
```

```python
pix_g3.shape
#차원 추가 유무 확인용
```

```python
pix_g=np.concatenate((pix_g1,pix_g2,pix_g3),axis=2)
pix_g.shape
```

```python
pix_g.dtype
```

```python
pix_g=np.round(pix_g)
```

```python
pix_g=pix_g.astype(int)
```

```python
plt.imshow(pix_g)
```

```python
pix_g=255-pix_g
# 127.5 기준으로 대칭점을 구하는 공식 = (pix값 + 원하는 값)/2=127.5 
# ex. pix 값이 255이면 0이어야 하고, 100이면 155이어야 127.5 기준 대칭이 완성
# 그래서, 계산 시 그냥 255- pix 값을 빼면 됨
```

```python
plt.imshow(pix_g)
```
