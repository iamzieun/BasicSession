```python
#정보요청 라이브러리 import
import requests

#크롤링 라이브러리 import
from bs4 import BeautifulSoup 
import pandas as pd
```

# 1

```python
#User-Agent 지정
headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.71 Safari/537.36'}
```

```python
#크롤링 할 대상 url 지정
url = 'https://www.saramin.co.kr/zf_user/jobs/list/domestic'

#크롤링 대상 url과 headers를 기반으로 정보 요청
res = requests.get(url, headers = headers)

#정상적으로 정보를 받은 경우 지속, 그렇지 않은 경우 중지
res.raise_for_status()

#html문서를 beautifulsoup의 lxml 형태로 변환
soup = BeautifulSoup(res.text, 'html.parser') 
```

```python
items = soup.select('div.col.company_nm > a > span')
```

```python
company_list = []
for item in items:
#     print(item.text)
    company_list.append(item.text)
```

```python
pd.DataFrame(company_list)
```

```python
#크롤링 할 대상 url 지정
url = 'https://www.saramin.co.kr/zf_user/jobs/list/domestic'

#크롤링 대상 url과 headers를 기반으로 정보 요청
res = requests.get(url, headers = headers)

#정상적으로 정보를 받은 경우 지속, 그렇지 않은 경우 중지
res.raise_for_status()

#html문서를 beautifulsoup의 lxml 형태로 변환
soup = BeautifulSoup(res.text, 'lxml') 
```

```python
#html문서에서 cluster_text_headline nclicks(cls_pol.clsart)라는 이름의 class를 지닌 a 태그의 element들 모두 선택
a1 = soup.findAll('a', attrs = {'class' : 'str_tit', "onclick":"s_trackApply(this, 'area_recruit', 'general');"})

#선택한 element들 내용 출력
a1
```

```python
extracted = list(map(lambda x:x.get_text().strip(), a1))
extracted
```

```python
name = extracted[0::2]
subject = extracted[1::2]
```

```python
df1 = pd.DataFrame({'name':name, 'subject':subject})
df1[:40]
```
