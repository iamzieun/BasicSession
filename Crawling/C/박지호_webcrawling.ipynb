```python
import requests
from bs4 import BeautifulSoup
import selenium
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(index = np.arange(1,41), columns = ['회사명','제목'])
```

```python
url1 = 'https://www.saramin.co.kr/zf_user/jobs/list/domestic'

res = requests.get(url1)
html = res.text
html = res1.text
soup = BeautifulSoup(html)
```

```python
#회사 추출
item_list =soup.find_all('div', class_ = 'list_item')
```

```python
#개수 조절
while len(item_list) > 40:
    item_list.pop(40)
    
print(len(item_list))
```

```python
#회사명, 제목 추출
n=1
for i in item_list:
    company_nm = i.find('div', class_ = 'col company_nm')
    company_nm = company_nm.find('span').get_text()
    tit = i.find('div', class_ = 'col notification_info')
    tit = tit.find('span').get_text()    
    df.loc[n] = [company_nm,tit]
    n +=1
```
