```python
import requests
from bs4 import BeautifulSoup 
import pandas as pd
```

```python
headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36'}
```

```python
list = []
```

```python
url = 'https://www.saramin.co.kr/zf_user/jobs/list/domestic'

res = requests.get(url, headers = headers)

res.raise_for_status()

soup = BeautifulSoup(res.text, 'lxml') 
```

```python
div = soup.findAll('div', attrs = {'class' : 'list_item'})
i = 0
for item in div:
    if (i >= 40):
        break
    item_dic = {}
    item_dic['기업명']=item.find(class_="company_nm").find(class_ = "str_tit")['title']
    item_dic['제목']=item.find(class_="job_tit").find(class_ = "str_tit")['title']
#     item_dic['지원자격']=item.find(class_="recruit_condition").get_text().strip()
#     item_dic['근무조건']=item.find(class_="company_info").get_text().strip()
#     item_dic['마감일*등록일']=item.find(class_="support_info").get_text().strip()
    list.append(item_dic)
    i+= 1 
```

```python
df = pd.DataFrame(list)
df
```
