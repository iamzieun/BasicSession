```python
#정보요청 라이브러리 import
import requests

#크롤링 라이브러리 import
from bs4 import BeautifulSoup 
```

```python
#User-Agent 지정
headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.71 Safari/537.36 Edg/97.0.1072.55'}
```

```python
#크롤링 할 대상 url 지정
url = 'https://www.saramin.co.kr/zf_user/jobs/list/domestic'

#크롤링 대상 url과 headers를 기반으로 정보 요청
res = requests.get(url, headers = headers)

#정상적으로 정보를 받은 경우 지속, 그렇지 않은 경우 중지
res.raise_for_status()

#html문서를 beautifulsoup의 lxml 형태로 변환
soup = BeautifulSoup(res.text, 'lxml') 
```

```python
#html문서에서 cluster_text_headline nclicks(cls_pol.clsart)라는 이름의 class를 지닌 a 태그의 element들 모두 선택
a1 = soup.findAll('a', attrs = {'class' : 'str_tit'})

#선택한 element들 내용 출력
a1
```

```python
import pandas as pd
import numpy as np
from pandas import DataFrame, Series
```

```python
something = Series(index = range(80))
```

```python
t=0
for a in a1:
    #a 태그의 element 내에서 href class attribute 정보 추출 및 링크 생성
    something[t] = a['title'] #위에서 확인할 수 있듯, href가 링크 값을 가짐.
    t = t+1
    if t >= 80:
        break
```

```python
something.head()
```

```python
employ = DataFrame(index = range(40), columns = ['company','name'])
```

```python
for t in range(80):
    if t % 2 == 0:
        employ['company'][t/2] = something[t]
    else:
        employ['name'][t//2]=something[t]
```
