```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup 
```

```python
url = 'https://www.saramin.co.kr/zf_user/jobs/list/domestic'
```

```python
req = requests.get(url, headers = headers)
req.raise_for_status()
soup = BeautifulSoup(req.text, 'lxml')
```

```python
div1 = soup.findAll('div', attrs = {'class':'list_item'})
div1
```

```python
names = []
titles = []

for div in div1:
    title = div.find('div', attrs={'class': 'col company_nm'}).a.get_text().strip()
    name = div.find('div', attrs={'class': 'col notification_info'}).a.get_text().strip()
    titles.append(name)
    names.append(title)
    if len(names) >= 40: break
```

```python
df = pd.DataFrame({'회사명':names,'내용':titles})
```
