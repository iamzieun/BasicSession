```python
import requests
from bs4 import BeautifulSoup
```

```python
headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36'}
```

```python
#크롤링 할 대상 url 지정
url = 'https://www.saramin.co.kr/zf_user/jobs/list/domestic'

#크롤링 대상 url과 headers를 기반으로 정보 요청
res = requests.get(url, headers = headers)

#정상적으로 정보를 받은 경우 지속, 그렇지 않은 경우 중지
res.raise_for_status()

#html문서를 beautifulsoup의 lxml 형태로 변환
soup = BeautifulSoup(res.text, 'lxml') 
```

```python
#html문서에서 cluster_text_headline nclicks(cls_pol.clsart)라는 이름의 class를 지닌 a 태그의 element들 모두 선택
a1 = soup.findAll('a', attrs = {'class' : 'str_tit'})

#선택한 element들 내용 출력
a1
```

```python
lst = []

#각 뉴스 제목과 링크를 추출할 반복문 생성
for a in a1:
    
    #a 태그의 element 내에서 title class attribute 정보 추출 및 링크 생성
    title = a['title']
    
    #title 출력
    print(f'title : {title}') 
    lst.append(title)
```

```python
company_name = []
announcement = []

for i in range(80):
  if i%2 == 0:
    company_name.append(lst[i])
  else:
    announcement.append(lst[i]) 
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(list(zip(company_name,announcement)), columns = ['회사명','제목'])

print(df)
```
